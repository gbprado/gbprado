```python
# conectar drive
from google.colab import drive
drive.mount('/content/drive')

import sqlite3

# conectar banco
conn = sqlite3.connect('/content/drive/MyDrive/Treinos_de_Python.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS Treinos_de_Python (
id INTEGER PRIMARY KEY AUTOINCREMENT,
nome TEXT NOT NULL,
email TEXT NOT NULL,
area TEXT NOT NULL,
idade TEXT NOT NULL
)
''')
conn.commit()

while True:
  print("\n==== Sistema de Usuários====")
  print("1 - Cadastrar usuário")
  print("2 - Ver usuários")
  print("3 - Atualizar usuário")
  print("4 - Deletar usuário")
  print("5 - Sair")

opcao = input("Escolha uma opção: ")

# CREATE
if opcao == "1":

        nome = input("Nome: ")
        email = input("Email: ")
        area = input("Área: ")
        idade = input("Idade: ")

        cursor.execute(
            "INSERT INTO Treinos_de_Python (nome,email,area,idade) VALUES (?,?,?,?)",
            (nome,email,area,idade)
        )

        conn.commit()
        print("Usuário cadastrado!")

# READ
elif opcao == "2":

        cursor.execute("SELECT * FROM Treinos_de_Python")
        dados = cursor.fetchall()

        print("\nID | Nome | Email | Área | Idade")
        print("----------------------------------")

        for linha in dados:
            print(linha[0], "|", linha[1], "|", linha[2], "|", linha[3], "|", linha[4])
# UPDATE
elif opcao == "3":

        id_usuario = input("Digite o ID do usuário que deseja atualizar: ")
        nova_area = input("Nova área: ")

        cursor.execute(
            "UPDATE Treinos_de_Python SET area = ? WHERE id = ?",
            (nova_area, id_usuario)
        )

        conn.commit()
        print("Usuário atualizado!")
# DELETE
elif opcao == "4":

        id_usuario = input("Digite o ID que deseja deletar: ")

        cursor.execute(
            "DELETE FROM Treinos_de_Python WHERE id = ?",
            (id_usuario,)
        )

        conn.commit()
        print("Usuário removido!")
# SAIR
elif opcao == "5":

        print("Encerrando sistema...")
        break

else:
        print("Opção inválida!")

conn.close()
```
